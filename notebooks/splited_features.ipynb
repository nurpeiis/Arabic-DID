```python
import os
import pandas as pd
```

```python
def init_count(a, folder):
    options = ['train', 'dev', 'test']
    a[folder] = dict()
    for option in options:
        a[folder][option] = 0
    return a

def init_set(a, folder):
    options = ['train', 'dev', 'test']
    a[folder] = dict()
    for option in options:
        a[folder][option] = set()
    return a

def append_set(parent_set, child_set):
    for child in child_set:
        parent_set.add(child)
    return parent_set
```

```python
parent_dir = '../../hierarchical-did/data_processed_second/'
folders = ['multi_dialect_multi_genre_informal_corpus_news', 'multi_dialect_multi_genre_informal_corpus_twitter', 'gumar']
word_count = dict()
char_count = dict()
word_set = dict()
char_set = dict()
lines_count = dict()
for folder in folders:
    word_count = init_count(word_count, folder)
    char_count = init_count(char_count, folder)
    word_set = init_set(word_set, folder)
    char_set = init_set(word_set, folder)
    lines_count = init_count(lines_count, folder)
    
    files = os.listdir(parent_dir+folder)
    print(f'Doing {folder}')
    
    for file in files:
        if 'info_data' in file or 'Store' in file:
            continue
        print(f'------- {file}')
        df = pd.read_csv(f'{parent_dir}{folder}/{file}', sep='\t', header=0)
        l_count = len(df.index)
        w_count = 0
        w_set = set()
        c_count = 0
        c_set = set()
        
        for index, row in df.iterrows():
            sentence = row['original_sentence']
            words = sentence.split(' ')
            for w in words:
                w_set.add(w)
            w_count += len(words)
            
            chars = list(sentence)
            for c in chars:
                c_set.add(c)
            c_count += len(chars)
        type_file = None
        if 'train' in file:
            type_file = 'train'
        elif 'dev' in file:
            type_file = 'dev'
        elif 'test' in file:
            type_file = 'test'
        
        lines_count[folder][type_file] += l_count
        word_count[folder][type_file] += w_count
        char_count[folder][type_file] += c_count
        word_set[folder][type_file] = append_set(word_set[folder][type_file], w_set)
        char_set[folder][type_file] = append_set(char_set[folder][type_file], c_set)
    
```

```python
options = ['train', 'dev', 'test']

def get_set_count(set_count):
    count = dict()
    
    for k in set_count.keys():
        count[k] = dict()
        total = 0
        for option in options:
            c = len(set_count[k][option])
            count[k][option] = c
            total += c
        count[k]['total'] = total
    
    return count
def get_total_count(count):
    for k in count.keys():
        total = 0
        for option in options:
            c = count[k][option]
            total += c
        count[k]['total'] = total 
    
    return count
```

```python
lines_count = get_total_count(lines_count)
word_count = get_total_count(word_count)
char_count = get_total_count(char_count)
word_set_count = get_set_count(word_set)
char_set_count = get_set_count(char_set)
```

```python
df = pd.read_csv('../datasets_splited_features.tsv', sep='\t', header=0)
df['dataset_name']
df.loc[17,'dataset_name']  = 'multi_dialect_multi_genre_informal_corpus_news'
df.loc[18,'dataset_name']  = 'multi_dialect_multi_genre_informal_corpus_twitter'
df
```

```python
def general_adder_to_df(df, name, count):
    lines_train = [0]*len(df)
    lines_dev = [0]*len(df)
    lines_test = [0]*len(df)
    lines_total = [0]*len(df)
    set_names = set(df['dataset_name'])
    for k in count.keys():
        if k not in set_names:
            continue
        i = df[df['dataset_name'] == k].index[0]
        lines_train[i] = count[k]['train']
        lines_test[i] = count[k]['test']
        lines_dev[i] = count[k]['dev']
        lines_total[i] = count[k]['total']
    
    df[f'{name}_train'] = lines_train
    df[f'{name}_test'] = lines_test
    df[f'{name}_dev'] = lines_dev
    df[f'{name}_total'] = lines_total
    
    return df
```

```python
df = general_adder_to_df(df, 'lines_count', lines_count)
df = general_adder_to_df(df, 'word_count', word_count)
df = general_adder_to_df(df, 'char_count', char_count)
df = general_adder_to_df(df, 'unique_word_count', word_set_count)
df = general_adder_to_df(df, 'unique_char_count', char_set_count)
df
```
